```python
class Municipio:
    def __init__(self, nombre, veterinarios):
        self.nombre = nombre
        self.veterinarios = veterinarios

class Veterinario:
    def __init__(self, nombre):
        self.nombre = nombre
        self.vacunas_aplicadas = 0

def agregar_municipio(municipios):
    nombre = input("Ingrese el nombre del municipio: ")
    veterinarios = []
    municipios.append(Municipio(nombre, veterinarios))
    print(f"Municipio: '{nombre}' agregado.")

def agregar_veterinario(municipio):
    nombre = input("Ingrese el nombre del veterinario: ")
    veterinario = Veterinario(nombre)
    municipio.veterinarios.append(veterinario)
    print(f"Veterinario '{nombre}' agregado al municipio '{municipio.nombre}'.")

def registrar_vacunas(municipio):
    print("Veterinarios en:", municipio.nombre)
    for idx, vet in enumerate(municipio.veterinarios, start=1):
        print(f"{idx}. {vet.nombre}")

    vet_idx = int(input("Seleccione el veterinario (número): ")) - 1
    if 0 <= vet_idx < len(municipio.veterinarios):
        cantidad = int(input("Ingrese el número de vacunas aplicadas: "))
        municipio.veterinarios[vet_idx].vacunas_aplicadas += cantidad
        print(f"Se registraron: {cantidad} vacunas para el veterinario: '{municipio.veterinarios[vet_idx].nombre}'.")
    else:
        print("Selección de veterinario incorrecta.")

def mostrar_municipios(municipios):
    print("Lista de municipios:")
    for idx, ciudad in enumerate(municipios, start=1):
        print(f"{idx}. {ciudad.nombre}")

def mostrar_total_vacunas_por_municipio(municipio):
    total_vacunas = sum(vet.vacunas_aplicadas for vet in municipio.veterinarios)
    print(f"Total de vacunas puestas en: {municipio.nombre}: {total_vacunas}")

def mostrar_total_vacunas_por_veterinario(municipio):
    print("Veterinarios en:", municipio.nombre)
    for idx, vet in enumerate(municipio.veterinarios, start=1):
        print(f"{idx}. {vet.nombre} - Total de vacunas: {vet.vacunas_aplicadas}")

def main():
    municipios = []

    while True:
        print("\n1. Agregar Municipio")
        print("2. Agregar Veterinario a Municipio")
        print("3. Registrar Vacunas para Veterinario en Municipio")
        print("4. Mostrar Municipios")
        print("5. Mostrar Total de Vacunas por Veterinario en Municipio")
        print("6. Mostrar Total de Vacunas por Municipio")
        print("7. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            agregar_municipio(municipios)
        elif opcion == "2":
            mostrar_municipios(municipios)
            municipio_idx = int(input("Seleccione el municipio (número): ")) - 1
            if 0 <= municipio_idx < len(municipios):
                agregar_veterinario(municipios[municipio_idx])
            else:
                print("Selección de municipio incorrecta.")
        elif opcion == "3":
            mostrar_municipios(municipios)
            municipio_idx = int(input("Seleccione el municipio (número): ")) - 1
            if 0 <= municipio_idx < len(municipios):
                registrar_vacunas(municipios[municipio_idx])
            else:
                print("Selección de municipio incorrecto.")
        elif opcion == "4":
            mostrar_municipios(municipios)
            municipio_idx = int(input("Seleccione el municipio (número): ")) - 1
            if 0 <= municipio_idx < len(municipios):
                mostrar_total_vacunas_por_veterinario(municipios[municipio_idx])
            else:
                print("Selección de municipio incorrecto.")
        elif opcion == "5":
            mostrar_municipios(municipios)
            municipio_idx = int(input("Seleccione el municipio (número): ")) - 1
            if 0 <= municipio_idx < len(municipios):
                mostrar_total_vacunas_por_municipio(municipios[municipio_idx])
            else:
                print("Selección de municipio inválida.")
        elif opcion == "6":
            mostrar_municipios(municipios)
        elif opcion == "7":
            break
        else:
            print("Opción incorrecta, intente nuevamente.")
main()
```
